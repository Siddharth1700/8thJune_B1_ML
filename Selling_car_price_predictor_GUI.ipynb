```python
#importing libraries
import pandas as pd
import tkinter as tk
from tkinter import *
from PIL import Image, ImageTk
import warnings
warnings.filterwarnings('ignore')
```

```python
m = pd.read_pickle('car_predictor.pkl')
```

```python
app = tk.Tk()

#image size=app size
app.geometry('800x500')
app.title('Car Selling Price Predictor Model')


v_year = tk.StringVar(app)
v_km_driven = tk.StringVar(app)
v_fuel = tk.StringVar(app)
v_seller_type = tk.StringVar(app)
v_transmission = tk.StringVar(app)
v_owner = tk.StringVar(app)
res = tk.Variable(app)



image = Image.open("car.jfif")

resize_image = image.resize((800, 500))
img = ImageTk.PhotoImage(resize_image)
label1 = Label(image=img)
label1.image = img
label1.pack()

def car_price():
    year=eval(field_year.get())
    km_driven=eval(field_km_driven.get())
    fuel=eval(field_fuel.get())
    seller_type=eval(field_seller_type.get())
    transmission=eval(field_transmission.get())
    owner=eval(field_owner.get())
    query=pd.DataFrame({'year':[year],'km_driven':[km_driven],'fuel':[fuel],'seller_type':[seller_type],'transmission':[transmission],'owner':[owner]})
    result.set(round(m.predict(query)[0],2))
    

            
label=tk.Variable(app)
tk.Label(app,textvariable=label,font=('Courier',15),fg='blue').pack()



Year=tk.Variable(app)
Year.set('Enter year of purchasing of car: ')
tk.Label(app,textvariable=Year,font=0).place(x=100,y=60)

Km_driven=tk.Variable(app)
Km_driven.set('Enter km driven: ')
tk.Label(app,textvariable=Km_driven,font=0).place(x=100,y=100)

Fuel=tk.Variable(app)
Fuel.set('Enter fuel type: ')
tk.Label(app,textvariable=Fuel,font=0).place(x=100,y=140)

Seller_type=tk.Variable(app)
Seller_type.set('Enter Seller Type: ')
tk.Label(app,textvariable=Seller_type,font=0).place(x=100,y=180)

Transmission=tk.Variable(app)
Transmission.set('Enter Transmission: ')
tk.Label(app,textvariable=Transmission,font=0).place(x=100,y=220)

Owner=tk.Variable(app)
Owner.set('Enter number of owner: ')
tk.Label(app,textvariable=Owner,font=0).place(x=100,y=260)

field_year=tk.Variable(app)
field_year.set('0')
tk.Entry(app,textvariable=field_year).place(x=500,y=60,height=20)

field_km_driven=tk.Variable(app)
field_km_driven.set('0')
tk.Entry(app,textvariable=field_km_driven).place(x=500,y=100,height=20)

field_fuel=tk.Variable(app)
field_fuel.set('0')
tk.Entry(app,textvariable=field_fuel).place(x=500,y=140,height=20)

field_seller_type=tk.Variable(app)
field_seller_type.set('0')
tk.Entry(app,textvariable=field_seller_type).place(x=500,y=180,height=20)

field_transmission=tk.Variable(app)
field_transmission.set('0')
tk.Entry(app,textvariable=field_transmission).place(x=500,y=220,height=20)

field_owner=tk.Variable(app)
field_owner.set('0')
tk.Entry(app,textvariable=field_owner).place(x=500,y=260,height=20)

tk.Button(app,text='Show selling price of car',command=car_price,bg='green',fg='black',font=0).place(x=250,y=300,height=40)

pricepred=tk.Variable(app)
pricepred.set('The predicted price of your car in Rs: ')
tk.Label(app,textvariable=pricepred,font=0).place(x=100,y=400)

result=tk.Variable(app)
result.set('')
tk.Label(app,textvariable=result,font=0).place(x=450,y=400)

app.mainloop()
```
